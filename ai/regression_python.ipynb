# Simple Linear Regression (Python)

This notebook demonstrates a simple linear regression analysis using Python to model **Salary** based on **Years of Experience**. We will load the data, visualize the relationship, fit a linear model, overlay the regression line, and evaluate model performance.

## Step 1: Load the Data

We read the CSV file into a pandas DataFrame and preview the first few rows to confirm the columns and values look correct.

```python
import pandas as pd

dataset = pd.read_csv("../regression_data.csv")
dataset.head()
```

## Step 2: Create a Scatter Plot

A scatter plot helps us visually inspect the relationship between years of experience (predictor) and salary (response).

```python
import matplotlib.pyplot as plt

plt.scatter(dataset["YearsExperience"], dataset["Salary"], color="red")
```

## Step 3: Fit a Simple Linear Regression Model

We use scikit-learn's `LinearRegression` to fit a model of the form:

$$\text{Salary} = \beta_0 + \beta_1 \times \text{YearsExperience}$$

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(dataset[["YearsExperience"]], dataset[["Salary"]])
```

## Step 4: Overlay the Regression Line

We plot the fitted line on top of the scatter plot to see how well the linear model captures the trend in the data.

```python
plt.plot(dataset["YearsExperience"], model.predict(dataset[["YearsExperience"]]), color="blue")
plt.title("Salary vs Experience")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()
```

## Step 5: Evaluate the Model

We use the **R-squared** ($R^2$) score to measure how much variance in salary is explained by years of experience. Values closer to 1 indicate a better fit.

```python
model.score(dataset[["YearsExperience"]], dataset[["Salary"]])
```
