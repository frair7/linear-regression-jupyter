# Simple Linear Regression (R)

This notebook demonstrates a simple linear regression analysis using R to model **Salary** based on **Years of Experience**. We will load the data, visualize the relationship, fit a linear model, overlay the regression line, and evaluate model performance.

## Step 1: Load the Data

We read the CSV file into an R data frame. The dataset contains years of experience and corresponding salary values.

```r
dataset <- read.csv("../regression_data.csv")
head(dataset)
```

## Step 2: Create a Scatter Plot

A scatter plot helps us visually inspect the relationship between years of experience (predictor) and salary (response).

```r
plot(dataset$YearsExperience, dataset$Salary, col = "red",
     xlab = "Years of Experience", ylab = "Salary",
     main = "Salary vs Experience")
```

## Step 3: Fit a Simple Linear Regression Model

We use R's `lm()` function to fit a model of the form:

$$\text{Salary} = \beta_0 + \beta_1 \times \text{YearsExperience}$$

```r
model <- lm(Salary ~ YearsExperience, data = dataset)
```

## Step 4: Overlay the Regression Line

Using ggplot2, we combine the scatter plot with the fitted regression line to visualize how well the model captures the trend.

```r
library(ggplot2)

ggplot() +
  geom_point(aes(x = dataset$YearsExperience, y = dataset$Salary), colour = "red") +
  geom_line(aes(x = dataset$YearsExperience, y = predict(model, newdata = dataset)), colour = "blue") +
  ggtitle("Salary vs Experience") +
  xlab("Years of Experience") +
  ylab("Salary")
```

## Step 5: Evaluate the Model

The model summary provides coefficient estimates, significance tests, and **Multiple R-squared** ($R^2$), which measures how much variance in salary is explained by years of experience.

```r
summary(model)
```
